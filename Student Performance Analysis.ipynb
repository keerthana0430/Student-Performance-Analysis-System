```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Student": ["Arun", "Priya", "Kavin", "Divya", "Rahul"],
    "Maths": [85, 90, 78, 88, 76],
    "Science": [92, 81, 74, 95, 80],
    "English": [75, 89, 84, 90, 70]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df["Average"] = (df["Maths"] + df["Science"] + df["English"]) / 3
```

```python
print(df[["Student", "Average"]])
```

```python
top_student = df.loc[df["Average"].idxmax()]

print("Top Performer:")
print(top_student["Student"])
print("Average Marks:", top_student["Average"])
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Student"], df["Average"])
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Performance Analysis")
plt.show()
```
